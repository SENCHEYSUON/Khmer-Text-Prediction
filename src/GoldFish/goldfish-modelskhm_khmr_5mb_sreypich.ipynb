```python
# ======================================================
# KHMER NEXT-WORD PREDICTION (FULL: TRAIN, PLOT, TEST, INFERENCE)
# ======================================================

import os
import re
import gc
import math
import random
import numpy as np
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
from datasets import load_dataset, Dataset, DatasetDict
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    DataCollatorForLanguageModeling,
    Trainer,
    TrainingArguments
)

# -----------------------------
# 1. Config & Paths
# -----------------------------
os.environ["WANDB_DISABLED"] = "true"
PROJECT_DIR = "/content/drive/MyDrive/Khmer_Next_Word_Prediction"
SAVE_PATH = os.path.join(PROJECT_DIR, "saved_model")
os.makedirs(SAVE_PATH, exist_ok=True)

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using Device: {device}")

MAX_INPUT = 64

# -----------------------------
# 2. Load & Process Dataset
# -----------------------------
print("Loading Khmer dataset...")
ds = load_dataset("seanghay/khPOS")

def clean_khmer_text(text):
    if text is None: return ""
    text = str(text).replace("_", " ").replace("~", "").replace("^", "").strip()
    text = re.sub(r"\s+", " ", text)
    return text

def build_examples(example):
    tokens = example.get("tokens", []) or []
    tokens = [clean_khmer_text(t) for t in tokens if t and str(t).strip()]
    rows = []
    for i in range(1, len(tokens)):
        context = " ".join(tokens[:i])
        target = tokens[i]
        if target in ["។", "!", "?", ".", "៖", "»", "«"]: continue
        rows.append({"context": context, "next_word": target})
    return rows

ds_trunc = ds["train"].select(range(min(4000, len(ds["train"]))))
processed = []
for ex in ds_trunc:
    processed.extend(build_examples(ex))

expanded_ds = Dataset.from_list(processed)
split_ds = expanded_ds.train_test_split(test_size=0.2, seed=SEED)
dataset = DatasetDict({"train": split_ds["train"], "validation": split_ds["test"]})

print(f"Training examples: {len(dataset['train'])}")
print(f"Validation examples: {len(dataset['validation'])}")

# -----------------------------
# 3. Model & Tokenizer
# -----------------------------
MODEL_NAME = "goldfish-models/khm_khmr_5mb"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)

special_tokens_dict = {}
if tokenizer.pad_token is None: special_tokens_dict["pad_token"] = "[PAD]"
if tokenizer.bos_token is None: special_tokens_dict["bos_token"] = "[SOS]"
if tokenizer.eos_token is None: special_tokens_dict["eos_token"] = "[EOS]"
if special_tokens_dict:
    tokenizer.add_special_tokens(special_tokens_dict)

model = AutoModelForCausalLM.from_pretrained(MODEL_NAME)
model.resize_token_embeddings(len(tokenizer))
model = model.to(device)
tokenizer.truncation_side = "left"

# -----------------------------
# 4. Tokenization
# -----------------------------
def tokenize_with_masking(batch):
    input_ids_list, labels_list = [], []
    for c, t in zip(batch["context"], batch["next_word"]):
        c_ids = tokenizer.encode(str(c or ""), add_special_tokens=False, truncation=True, max_length=MAX_INPUT-5)
        t_ids = tokenizer.encode(" " + str(t or ""), add_special_tokens=False)
        full_ids = ([tokenizer.bos_token_id] if tokenizer.bos_token_id else []) + c_ids + t_ids + ([tokenizer.eos_token_id] if tokenizer.eos_token_id else [])
        labels = ([-100]*(len(c_ids)+ (1 if tokenizer.bos_token_id else 0)) + t_ids + ([tokenizer.eos_token_id] if tokenizer.eos_token_id else []))[:MAX_INPUT]
        full_ids = full_ids[:MAX_INPUT]
        input_ids_list.append(full_ids)
        labels_list.append(labels)

    enc = tokenizer.pad({"input_ids": input_ids_list}, padding="max_length", max_length=MAX_INPUT, return_tensors="pt")
    lab = tokenizer.pad({"input_ids": labels_list}, padding="max_length", max_length=MAX_INPUT, return_tensors="pt")
    enc["labels"] = torch.tensor([[-100 if x == tokenizer.pad_token_id else x for x in row] for row in lab["input_ids"]])
    return enc

tokenized_ds = dataset.map(tokenize_with_masking, batched=True, batch_size=100, remove_columns=dataset["train"].column_names)
data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)

# -----------------------------
# 5. Metrics
# -----------------------------
def preprocess_logits_for_metrics(logits, labels):
    return torch.argmax(logits, dim=-1)

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    mask = (labels != -100)
    correct = (predictions == labels) & mask
    accuracy = correct.sum() / (mask.sum() + 1e-9)
    return {"accuracy": accuracy.item()}

# -----------------------------
# 6. Training
# -----------------------------
training_args = TrainingArguments(
    output_dir=PROJECT_DIR,
    num_train_epochs=30,
    learning_rate=5e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_checkpointing=True,
    eval_accumulation_steps=1,
    fp16=torch.cuda.is_available(),
    eval_strategy="epoch",
    save_strategy="epoch",
    logging_steps=50,
    report_to="none",
    load_best_model_at_end=True,
    metric_for_best_model="accuracy"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_ds["train"],
    eval_dataset=tokenized_ds["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    preprocess_logits_for_metrics=preprocess_logits_for_metrics
)

print("\nStarting Training...")
trainer.train()

# -----------------------------
# 7. Save Model & Tokenizer
# -----------------------------
print(f"\nSaving model to: {SAVE_PATH}")
trainer.save_model(SAVE_PATH)
tokenizer.save_pretrained(SAVE_PATH)

# -----------------------------
# 8. Plot Metrics & PERFORMANCE TABLE
# -----------------------------
logs = trainer.state.log_history
epochs, train_losses, eval_losses, perplexities, accuracies = [], [], [], [], []

for l in logs:
    if "eval_loss" in l:
        epochs.append(l["epoch"])
        eval_losses.append(l["eval_loss"])
        perplexities.append(math.exp(l["eval_loss"]))
        accuracies.append(l.get("eval_accuracy", 0))
    if "loss" in l:
        train_losses.append(l["loss"])

# --- MANUAL PERFORMANCE TABLE ---
print("\n" + "="*80)
print(f"{'Epoch':^8} | {'Val Loss':^12} | {'Perplexity (e^loss)':^22} | {'Accuracy %':^12}")
print("-" * 80)
for i in range(len(eval_losses)):
    print(f"{epochs[i]:^8.1f} | {eval_losses[i]:^12.4f} | {perplexities[i]:^22.2f} | {accuracies[i]*100:^11.2f}%")
print("="*80)

plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
plt.plot(eval_losses, label="Validation Loss", marker='o', color='red')
plt.title("Loss Trend"); plt.legend()

plt.subplot(1, 3, 2)
plt.plot(perplexities, label="Perplexity", marker='s', color='orange')
plt.title("Perplexity (Lower is better)"); plt.legend()

plt.subplot(1, 3, 3)
plt.plot(accuracies, label="Accuracy", marker='^', color='green')
plt.title("Accuracy Trend"); plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------
# 9. Inference / Test (PREFIX TABLE)
# -----------------------------
def infer_next_words_table(prefix_list, top_k=3):
    model.eval()
    print("\n" + "="*85)
    print(f"{'Prefix Context':<40} | {'Top Predictions (Word: Prob)':<40}")
    print("-" * 85)
    for prefix in prefix_list:
        inputs = tokenizer(prefix, return_tensors="pt", truncation=True, max_length=MAX_INPUT).to(device)
        with torch.no_grad():
            logits = model(**inputs).logits[0, -1]
        for sid in tokenizer.all_special_ids: logits[sid] = -float("inf")
        probs = F.softmax(logits, dim=-1)
        topv, topi = torch.topk(probs, top_k)

        preds = ", ".join([f"{tokenizer.decode([idx]).strip()}: {v.item()*100:.1f}%" for v, idx in zip(topv, topi)])
        # Truncate prefix for display if too long
        display_prefix = (prefix[:37] + '...') if len(prefix) > 37 else prefix
        print(f"{display_prefix:<40} | {preds:<40}")
    print("="*85)

test_cases = [
    "រាជ រដ្ឋាភិបាល កម្ពុជា បាន", "ក្រសួង សុខាភិបាល បាន", "ក្រសួង អប់រំ យុវជន និង កីឡា បាន",
    "យោង តាម សេចក្តី ប្រកាស របស់", "យោង តាម ការ លើក ឡើង របស់", "ក្នុង ករណី នេះ ត្រូវ",
    "សេចក្តី ប្រកាស នេះ មាន", "អាជ្ញាធរ មាន សមត្ថកិច្ច បាន", "គណៈ រដ្ឋមន្ត្រី បាន", "ស្ថាប័ន នេះ មាន ភារកិច្ច"
]
infer_next_words_table(test_cases)

```

```python
# -----------------------------
# 10. Save Validation Predictions
# -----------------------------
def save_test_results(dataset, save_file):
    model.eval()
    with open(save_file, "w", encoding="utf-8") as f:
        for row in dataset:
            context = row["context"]
            true_word = row["next_word"]
            inputs = tokenizer(context, return_tensors="pt", truncation=True, max_length=MAX_INPUT).to(device)
            with torch.no_grad():
                logits = model(**inputs).logits[0, -1]
            for sid in tokenizer.all_special_ids: logits[sid] = -float("inf")
            pred_id = torch.argmax(logits).item()
            pred_word = tokenizer.decode([pred_id]).strip()
            f.write(f"Context: {context}\nTrue: {true_word}\nPred: {pred_word}\n{'-'*50}\n")
    print(f"Validation predictions saved to: {save_file}")

save_test_results(dataset["validation"],
                  os.path.join(PROJECT_DIR, "validation_predictions.txt"))
```
